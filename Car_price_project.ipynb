```python
#we need to only encode name,location,fuel_type,tranmission,owner_type
#the rest of the object column is actually integer but since the unit is in string it came as object
```

```python
# GetDummies encoding

df1 = pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df1
```

```python
# concatenating encoded dataframe and original dataframe
df3 = pd.concat([df,df1],axis=1)
df3
```

```python
#since name has 1878 rows we are dropping Name for the time being
#we should drop unnamed0 column because it doesnt provide any value
#we are dropping new_price column since it has 5000+ Na values where total rows is 6000
```

```python
#dropping unnecessary columns
df3 = df3.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1)
```

```python
#replacing string in mileage
df3['Mileage'] = df3['Mileage'].str.replace('km/kg','')
df3['Mileage'] = df3['Mileage'].str.replace('kmpl','')
```

```python
#replacing string in engine
df3['Engine'] = df3['Engine'].str.replace('CC','')
```

```python
#replacing string in power
df3['Power'] = df['Power'].str.replace('bhp','')
```

```python
df3['Mileage']=df3['Mileage'].str.replace('null','0') #why 0 is in quotes because it is a string column
df3['Engine']=df3['Engine'].str.replace('null','0')
df3['Power']=df3['Power'].str.replace('null','0')
```

```python
#converting object column to float
df3['Mileage'] = df3['Mileage'].astype(float)
df3['Engine']=df3['Engine'].astype(float)
df3['Power'] = df3['Power'].astype(float)
```

```python
df3.dtypes
```

```python
#converting zeros to NaN
df3.loc[df3.Mileage==0,'Mileage'] = np.NaN
df3.loc[df3.Power==0,'Power'] =np.NaN
df3.loc[df3.Engine==0,'Engine'] = np.NaN
```

```python
#filling missing values
df3['Mileage'].fillna(df3['Mileage'].mean(),inplace=True)
df3['Engine'].fillna(df3['Engine'].mean(),inplace=True)
df3['Power'].fillna(df3['Power'].mean(),inplace=True)
df3['Seats'].fillna(df3['Seats'].mode()[0],inplace=True)
df3.isna().sum()
```

```python
#seperating it into x and y
x=df3.drop(['Price'],axis=1)
y=df3['Price']
```
